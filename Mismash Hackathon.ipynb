```python
x=train[['Social_Search_Impressions', 'Social_Search_Working_cost',
       'Digital_Impressions', 'Digital_Working_cost',
       'Print_Impressions.Ads40', 'Print_Working_Cost.Ads50',
       'OOH_Impressions', 'OOH_Working_Cost', 'SOS_pct',
       'Digital_Impressions_pct', 'CCFOT', 'Median_Temp', 'Median_Rainfall',
       'Fuel_Price', 'Inflation', 'Trade_Invest', 'Brand_Equity',
       'Avg_EQ_Price', 'Any_Promo_pct_ACV', 'Any_Feat_pct_ACV',
       'Any_Disp_pct_ACV', 'EQ_Base_Price', 'Est_ACV_Selling', 'pct_ACV',
       'Avg_no_of_Items', 'pct_PromoMarketDollars_Category', 'RPI_Category',
       'Magazine_Impressions_pct', 'TV_GRP', 'Competitor1_RPI',
       'Competitor2_RPI', 'Competitor3_RPI', 'Competitor4_RPI', 'EQ_Category',
       'EQ_Subcategory', 'pct_PromoMarketDollars_Subcategory',
       'RPI_Subcategory']]
y=train['EQ']
print(x.shape)
```

```python
from sklearn.model_selection import KFold, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import SCORERS, mean_absolute_error,mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
```

```python
lr=LinearRegression(normalize=True)
model_lr=lr.fit(x_train,y_train)
pred_lr=lr.predict(x_test)
mean_absolute_error(y_test,pred_lr)
```

```python
rf=RandomForestRegressor()
model_rf=rf.fit(x_train,y_train)
pred_rf=rf.predict(x_test)
print(mean_absolute_error(y_test,pred_rf))
print(mean_squared_error(y_test,pred_rf))
```

```python
model = ExtraTreesRegressor()
model.fit(x_train,y_train)
```

```python
print(model.feature_importances_)
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=x_train.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest, RFECV, f_regression,mutual_info_regression
```

```python
selected_feature=SelectKBest(score_func=f_regression,k=15).fit(x_train,y_train)
df1=pd.DataFrame({'features':list(x_train.columns),'scores':selected_feature.scores_})
df1.sort_values(by='scores',ascending=False)
```

```python
df1=pd.DataFrame({'features':list(x_train.columns),'scores':selected_feature.scores_})
df1.sort_values(by='scores',ascending=False)
```

```python
x_train_re=selected_feature.transform(x_train)
x_test_re=selected_feature.transform(x_test)
```

```python
model_re=rf.fit(x_train_re,y_train)
pred_res=rf.predict(x_test_re)
print(mean_absolute_error(y_test,pred_res))
print(mean_squared_error(y_test,pred_res))
```

```python
selected_features=SelectKBest(score_func=mutual_info_regression,k=10).fit(x_train,y_train)
df2=pd.DataFrame({'features':list(x_train.columns),'scores':selected_features.scores_})
df2.sort_values(by='scores',ascending=False)
```

```python
x_train_re=selected_features.transform(x_train)
x_test_re=selected_features.transform(x_test)
```

```python
model_re=rf.fit(x_train_re,y_train)
pred_re=rf.predict(x_test_re)
```

```python
mean_absolute_error(y_test,pred_re)
```

```python
rfe=RFECV(estimator=lr,scoring='neg_mean_absolute_error',step=1,cv=10)
rfe.fit(x_train,y_train)
```

```python
print('optimal features',rfe.n_features_)
print('best features',x_train.columns[rfe.support_])
```

```python
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score of number of selected features")
plt.plot(range(1, len(rfe.grid_scores_) + 1), rfe.grid_scores_)
plt.show()
```

```python
x_train_rf=rfe.transform(x_train)
x_test_rf=rfe.transform(x_test)
```

```python
model_rf=rf.fit(x_train_rf,y_train)
pred_rf=rf.predict(x_test_rf)
```
